```python
def city(a,b,c,d):
    
    b.drop('Year',axis=1,inplace=True)
    df = pd.concat([a,b,c,d],axis=0)
    df.set_index('Month',inplace=True)
    df = df.set_index(pd.Index(['Avg Temp Min(°C)', 'Avg Temp Max(°C)', 'Avg Temp Min(°F)', 'Avg Temp Max(°F)','Avg Prec.(mm)','Avg Prec.(in)','No. days of Prec','Avg Hours Sunlight','Sea Temp(°C)','Sea Temp(°F)']))
    df.drop(['Avg Temp Min(°F)', 'Avg Temp Max(°F)', 'Sea Temp(°F)'], inplace=True)
    return df
```

```python
capetown = city(df[0],df[1],df[2],df[3])
```

```python
capetown_df = capetown.transpose()
capetown_df['Month']  = ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']
capetown_df.head()
```

```python
portelizabeth = city(df[13],df[12],df[14],df[15])
```

```python
portelizabeth_df = portelizabeth.transpose()
portelizabeth_df['Month']  = ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']
portelizabeth_df.head()
```

```python
durban = city(df[17],df[16],df[18],df[19])
durban
```

```python
durban_df = durban.transpose()
durban_df['Month']  = ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']
durban_df.head()
```

```python
import dash
import dash_core_components as dcc
import dash_html_components as html
import plotly.graph_objs as go
import pandas as pd

# Launch the application:
app = dash.Dash()

# Create a Dash layout that contains a Graph component:
app.layout = html.Div(children = [
    html.H1('WEATHER FOR 3 COSTAL CITIES'),
    dcc.Graph(id='temperature',
             figure ={
                 'data': [
                     {'x':portelizabeth_df['Month'],'y':portelizabeth_df['Avg Temp Max(°C)'],'type':'line','name':'Port Elizabeth'},
                     {'x':durban_df['Month'],'y':durban_df['Avg Temp Max(°C)'],'type':'line','name':'Durban'},
                     {'x':capetown_df['Month'],'y':capetown_df['Avg Temp Max(°C)'],'type':'line','name':'Cape Town'}
                 ],
                 'layout':{
                     'title': 'Maximum Average Temperature'
                 }
             }),
    html.Div(children='''Overall we can see that out of all three cities Durban is considerably hotter. 
    Especially during autum/winter it is much more hotter. Cape Town is hotter than PE except for autum/winter,
    when PE is hotter. This tells us that Cape Town is hot during summer and the gets very cold during winter.'''),
    dcc.Graph(id='temperature1',
             figure ={
                 'data': [
                     {'x':portelizabeth_df['Month'],'y':portelizabeth_df['Avg Prec.(mm)'],'type':'bar','name':'Port Elizabeth'},
                     {'x':durban_df['Month'],'y':durban_df['Avg Prec.(mm)'],'type':'bar','name':'Durban'},
                     {'x':capetown_df['Month'],'y':capetown_df['Avg Prec.(mm)'],'type':'bar','name':'Cape Town'}
                 ],
                 'layout':{
                     'title': 'Average Precipitation.(mm)'
                 }
             }),
    html.Div(children='''Durban has the highest amount of precipitation and PE has the lowest. One thing that really
    stands out is that CPT has a much greater amount of precipitation during winter when it has really cold temperatures. Durban is the
    opposite and has it's higest levels of precipitation during spring/summer when it has really hot temperatures.
    PE has the least amount of precipitation. With the lowest levels of precipitation being during summer. 
    Other than that there aren't any dramatic changes for PE. CPT and PE preipitaion seem to be more inversly
    proportinal to the weather whereas Durban is more directly proportional.'''),
    dcc.Graph(id='temperature2',
             figure ={
                 'data': [
                     {'x':portelizabeth_df['Month'],'y':portelizabeth_df['No. days of Prec'],'type':'bar','name':'Port Elizabeth'},
                     {'x':durban_df['Month'],'y':durban_df['No. days of Prec'],'type':'bar','name':'Durban'},
                     {'x':capetown_df['Month'],'y':capetown_df['No. days of Prec'],'type':'bar','name':'Cape Town'}
                 ],
                 'layout':{
                     'title': 'No. days of Precipitation'
                 }
             }),
    html.Div(children='''The number of days of precipitation is consistent with the amount if precipitation.
    With Durban having the most, then CPT and lastly PE'''),
    dcc.Graph(id='temperature3',
             figure ={
                 'data': [
                     {'x':portelizabeth_df['Month'],'y':portelizabeth_df['Avg Hours Sunlight'],'type':'line','name':'Port Elizabeth'},
                     {'x':durban_df['Month'],'y':durban_df['Avg Hours Sunlight'],'type':'line','name':'Durban'},
                     {'x':capetown_df['Month'],'y':capetown_df['Avg Hours Sunlight'],'type':'line','name':'Cape Town'}
                 ],
                 'layout':{
                     'title': 'Average hours of Sunlight'
                 }
             }),
    html.Div(children='''CPT has really long days during summer and very short days during winter. So the colder it
    gets in CPT the shorter the days become. The same goes for PE. However, whats really interesting is to see that
    the opposite is true for Durban. The colder it gets the longer the days seem to be. '''),
    dcc.Graph(id='temperature4',
             figure ={
                 'data': [
                     {'x':portelizabeth_df['Month'],'y':portelizabeth_df['Sea Temp(°C)'],'type':'line','name':'Port Elizabeth'},
                     {'x':durban_df['Month'],'y':durban_df['Sea Temp(°C)'],'type':'line','name':'Durban'},
                     {'x':capetown_df['Month'],'y':capetown_df['Sea Temp(°C)'],'type':'line','name':'Cape Town'}
                 ],
                 'layout':{
                     'title': 'Average Sea temperature'
                 }
             }),
    html.Div(children='''From this we can see that the sea temperature is dependant on the overall temperature of
    the city. However it seems like there are other factors like location that have an impact on the sea temperature
    as CPT is hotter than PE in summer but it's sea temperature is colder than PE '''),
    html.H1('CONCLUSION'),
    html.Div(children='''Overall we can see that Cape Town has extremes whereby it is really cold in winter then
    really hot during summer. Also in terms of Precipitation it has very little during summer and a lot more during 
    winter. The same goes for Amount of Sunlight, it has really long days during summer and very short ones during
    summer. CPT's sea temperature does't seem to have and extremely dramtic changes. PE doesn'have any extremely 
    dramatic changes. Nor does Durban except that the amount of precipitation is a lot less during winter than in summer. ''')    
    
])

# Add the server clause:
if __name__ == '__main__':
    app.run_server()
```
